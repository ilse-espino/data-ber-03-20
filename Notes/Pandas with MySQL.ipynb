```python
order_items = pd.read_sql("SELECT * FROM order_items", conn)
products = pd.read_sql("SELECT * FROM products", conn)
```

```python
order_items.dtypes
```

```python
products.dtypes
```

```python
order_products = order_items.merge(products,
                 how="left",
                 on="product_id")[["order_id",
                                 "order_item_id",
                                 "product_id",
                                 "price",
                                 "product_category_name"]]

order_products
```

```python
per_category = order_products.groupby("product_category_name").agg({"price": "sum",
                                                    "order_id": "count"})

per_category
```

```python
per_category.reset_index()
```

```python
# all together

(order_items
.merge(products, how="left", on="product_id")
.groupby("product_category_name")
.agg({"price":"sum",
     "order_id":"count"})
.reset_index())
```

```python
# add a column "javier"
# rename columns

favorite_df = (order_items
                    .merge(products, how="left", on="product_id")
                    .groupby("product_category_name")
                    .agg({"price":"sum",
                         "order_id":"count"})
                    .reset_index()
                    .rename(columns={"price":"revenue",
                                    "order_id":"items_sold"})
                    .assign(javier=1))
```

```python
# sorting

(order_items
.merge(products, how="left", on="product_id")
.groupby("product_category_name")
.agg({"price":"sum",
"order_id":"count"})
.reset_index()
.rename(columns={"price":"revenue",
                  "order_id":"items_sold"})
.sort_values("revenue", ascending=False)).head(10)
```
