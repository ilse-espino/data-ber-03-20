## Challenge 2: Count Commits

Count how many commits were made in the past week.

1. Obtain all the commits made in the past week via API, which is a JSON array that contains multiple commit objects.

```python
url = "https://api.github.com/repos/ironhack-datalabs/data-ber-03-20/commits"
headers = {"Accept": "application/vnd.github.v3+json"}
resp = requests.get(url, headers)
resp
```

```python
commits = resp.json()
```

```python
commits
```

```python
commits_df = pd.DataFrame(commits)

commits_df["commit"][0]["author"]["date"][:10]
```

```python
dates_lst = [commits_df["commit"][i]["author"]["date"][:10] for i in range(len(commits_df))]

len(dates_lst)
```

```python
commits_df["dates"] = dates_lst
```

```python
commits_df["dates"] = pd.to_datetime(commits_df["dates"])

commits_df.dtypes
```

```python
today = datetime.date.today() 
week_ago = today - datetime.timedelta(days=6)
```

```python
commits_df["week_ago"] = week_ago.strftime("%Y-%m-%d")
commits_df["week_ago"] = pd.to_datetime(commits_df["week_ago"])
commits_df.dtypes

```

```python
commits_week = commits_df[commits_df["week_ago"] <= commits_df["dates"]]

commits_week
```
